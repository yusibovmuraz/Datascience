```python
#Q8-Write a class named arithmetic with method for calculating addition, substraction, multiplication and division of two numbers.
class arithmetic:
  def addition(x,y):
    return x+y
print(arithmetic.addition(8,5))

class arithmetic:
  def substraction(x,y):
    return x-y
print(arithmetic.substraction(8,5))

class arithmetic:
  def multiplication(x,y):
    return x*y
print(arithmetic.multiplication(8,5))

class arithmetic:
  def division(x,y):
    return x/y
print(arithmetic.division(8,5))
```
